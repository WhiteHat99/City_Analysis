```python
#Importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# ANALYZE AND SHARE
### ANSWERING THE QUESTIONS IN THE ASK SEGMENT

### 1. WHICH NEIGHBOURHOODS HAVE THE HIGHEST AVERAGE PRICES IN BOTH CITIES?
Based on the visualization below; Centrum-West, Centrum-Oost and Zuid are the most expensive neighbourhoods in Amsterdam with an average price of $189.92, $189.38 and $171.13 respectively while Noho, Tribeca and Soho are the most expensive neighbourhoods in New York with an average price of $304, $301.21 and $296.18 respectively.

```python
# Average price for each neighbourhood in Amsterdam
Amsterdam_average = Amsterdam.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10).round(2)
print('Average price for neighbourhood in Amsterdam:\n', Amsterdam_average)

# Average price for each neighbourhood in New York
New_York_average = New_York.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10).round(2)
print('Average price for neighbourhood in New York:\n', New_York_average)

# Plotting
plt.figure(figsize=(10, 8))

# Bar chart for Amsterdam average price
plt.subplot(1, 2, 1)
Amsterdam_average.plot(kind='bar', color='dodgerblue')
plt.title('Average Amsterdam Neighbourhood Price')
plt.ylabel('Average Price')

# Bar chart for New_York average price
plt.subplot(1, 2, 2)
New_York_average.plot(kind='bar', color='mediumseagreen')
plt.title('Average New York Neighbourhood Price')
plt.ylabel('Average Price')

plt.tight_layout()
plt.show()
```

### 2. WHICH NEIGBOURHOODS HAVE THE HIGHEST NUMBER OF LISTINGS?
Based on the visualization below; De Baarsjes - Oud-West, De Pijp - Rivierenbuurt and Centrum-West has the highest neighbourhood listings of 3515, 2493 and 2326 properties respectively for Amsterdam while Brooklyn, Manhattan and Queens has the highest neigbourhood listings of 3875, 3229 and 1099 properties respectively for New York.

```python
# Highest number of listings for each neighbourhood in Amsterdam
Amsterdam_list = Amsterdam['neighbourhood'].value_counts().sort_values(ascending=False).head(10)
print('Amsterdam Neighbourhood Listings:\n', Amsterdam_list)

# Highest number of listings for each neighbourhood in New York
New_York_list = New_York['neighbourhood'].value_counts().sort_values(ascending=False).head(10)
print('New York Neighbourhood Listings:\n', New_York_list)

# Plotting
plt.figure(figsize=(10, 8))

# Bar chart for Amsterdam listings
plt.subplot(1, 2, 1)
Amsterdam_list.plot(kind='bar', color='dodgerblue')
plt.title("Top 10 Amsterdam Neighbourhood Listings")
plt.ylabel("Number of Listings")

# Bar chart for New_York listings
plt.subplot(1, 2, 2)
New_York_list.plot(kind='bar', color='mediumseagreen')
plt.title("Top 10 New York Neighbourhood Listings")
plt.ylabel("Number of Listings")

plt.tight_layout()
plt.show()
```

### 3. WHICH ROOM TYPE IS MOST COMMON IN EACH CITY?
Based on the visualization below, Entire home/apt is the most common room type with 15889 listings in Amsterdam. Also, Entire home/apt is the most common room type with 9405 listings in New York.

```python
# Most common room type in Amsterdam
Amsterdam_room = Amsterdam['room_type'].value_counts().sort_values(ascending=False)
print('Amsterdam Room Types:\n', Amsterdam_room)

# Most common room type in New York
New_York_room = New_York['room_type'].value_counts().sort_values(ascending=False)
print('New York Room Types:\n', New_York_room)

# Plotting
plt.figure(figsize=(10, 5))

# Bar chart for Amsterdam room types
plt.subplot(1, 2, 1)
Amsterdam_room.plot(kind='bar', color='dodgerblue')
plt.title("Amsterdam Room Types")
plt.ylabel("Number of Listings")

# Bar chart for New_York room types
plt.subplot(1, 2, 2)
New_York_room.plot(kind='bar', color='mediumseagreen')
plt.title('New York Room Types')
plt.ylabel("Number of Listings")

plt.tight_layout()
plt.show()
```

### 4. WHICH CITY HAS CHEAPER PRIVATE ROOMS AND ENTIRE HOMES ON AVERAGE?
Based on the visualization below, New York has a cheaper private room with an average price of $82.46 and Amsterdam has a cheaper entire homes with an average price of $165.78.

```python
# Calculate average prices by room type for both cities
Avg_price_amsterdam = Amsterdam.groupby("room_type")["price"].mean().round(2)
Avg_price_new_york = New_York.groupby("room_type")["price"].mean().round(2)

# Filter only "Private room" and "Entire home/apt"
Amsterdam_avg_price = Avg_price_amsterdam[["Private room", "Entire home/apt"]]
New_York_avg_price = Avg_price_new_york[["Private room", "Entire home/apt"]]

# Print the average prices
print('Average Room Price for Amsterdam:\n', Amsterdam_avg_price)
print('Average Room Price for New York:\n', New_York_avg_price)

# Plotting
plt.figure(figsize=(8, 5))

# Combine into one DataFrame for plotting
Avg_prices = pd.DataFrame({'Amsterdam': Amsterdam_avg_price,
                            "New York": New_York_avg_price})

# Plot
Avg_prices.plot(kind='bar', color=['dodgerblue', 'mediumseagreen'])
plt.title('Average Price of Private Rooms and Entire Homes')
plt.ylabel('Average Price')
plt.xlabel('Room Type')
plt.legend(title='City')

plt.tight_layout()
plt.show()
```

### 5. WHAT IS THE AVERAGE PRICE OF LISTINGS IN AMSTERDAM VS NEW YORK?
Based on the visualization below, Amsterdam has the average listing price of $152.18 while New York has the average listing price of $137.46.

```python
# Calculate average price per city
avg_price_amsterdam = Amsterdam['price'].mean().round(2)
avg_price_newyork = New_York['price'].mean().round(2)

# Put into DataFrame for plotting
avg_prices = pd.DataFrame({'City': ['Amsterdam', 'New York'],
                        'Average Price': [avg_price_amsterdam, avg_price_newyork]})

# Print the average prices
print('Average Listing Price for Amsterdam:\n', avg_price_amsterdam)
print('Average Listing Price for New York:\n', avg_price_newyork)

# Plot bar chart
plt.figure(figsize=(8,5))
plt.bar(avg_prices['City'], avg_prices['Average Price'], color=['dodgerblue', 'mediumseagreen'])
plt.title('Average Listing Price: Amsterdam vs New York')
plt.ylabel('Average Price')
plt.xlabel('City')

plt.tight_layout()
plt.show()
```

### 6. HOW DO THE MINIMUM NIGHTS REQUIRED COMPARE BETWEEN AMSTERDAM AND NEW YORK?
Based on the visualization below, New York's average is higher than Amsterdam's which shows that Amsterdam is friendlier for short stays.

```python
# Calculate average minimum nights per city
Amsterdam_min_nights = Amsterdam['minimum_nights'].mean().round(2)
New_York_min_nights = New_York['minimum_nights'].mean().round(2)

# Print the Cities min nights
print('Amsterdam Minimum Nights:\n', Amsterdam_min_nights)
print('New York Minimum Nights:\n', New_York_min_nights)

# Put into DataFrame for plotting
Min_nights = pd.DataFrame({'City': ['Amsterdam', 'New York'],
                        'Average Minimum Nights': [Amsterdam_min_nights, New_York_min_nights]})

# Bar plot
plt.figure(figsize=(8,5))
plt.bar(Min_nights['City'], Min_nights['Average Minimum Nights'], color=['dodgerblue', 'mediumseagreen'])
plt.title('Average Minimum Nights Required: Amsterdam vs New York')
plt.ylabel('Average Minimum Nights')
plt.xlabel('City')

plt.tight_layout()
plt.show()
```

### 7. WHAT IS THE AVERAGE NUMBER OF REVIEWS PER LISTING IN EACH CITY?
Based on the visualization, Amsterdam has 21.56 reviews per listing while New York has 56.13 reviews per listings.

```python
# Calculate average number of reviews per city
Amsterdam_reviews = Amsterdam['number_of_reviews'].mean().round(2)
New_York_reviews = New_York['number_of_reviews'].mean().round(2)

# Print the cities review numbers
print('Number of Reviews for Amsterdam:\n', Amsterdam_reviews)
print('Number of Reviews for New York:\n', New_York_reviews)

# Put into DataFrame for plotting
Reviews = pd.DataFrame({'City': ['Amsterdam', 'New York'],
                        'Average Number of Reviews': [Amsterdam_reviews, New_York_reviews]})

# Bar plot
plt.figure(figsize=(8,5))
plt.bar(Reviews['City'], Reviews['Average Number of Reviews'], color=['dodgerblue', 'mediumseagreen'])
plt.title('Average Number of Reviews: Amsterdam vs New York')
plt.ylabel('Average Number of Reviews')
plt.xlabel('City')

plt.tight_layout()
plt.show()
```

### 8. HOW DOES THE AVAILABILITY OF LISTINGS DIFFER BETWEEN AMSTERDAM AND NEW YORK?
Based on the visualization below, New York has the most availability listing of 154.15 over Amsterdam which has 59.91.

```python
# Calculate average availability per city
Amsterdam_availability = Amsterdam['availability_2019'].mean().round(2)
New_York_availability = New_York['availability_2020'].mean().round(2)

# Print the cities availability
print('Amsterdam Average Availability:\n', Amsterdam_availability)
print('New York Average Availability:\n', New_York_availability)

# Put into DataFrame for plotting
Avg_availability = pd.DataFrame({'City': ['Amsterdam', 'New York'],
                                'Average Availability': [Amsterdam_availability, New_York_availability]})

# Bar plot
plt.figure(figsize=(8,5))
plt.bar(Avg_availability['City'], Avg_availability['Average Availability'], color=['dodgerblue', 'mediumseagreen'])
plt.title('Average Availability: Amsterdam vs New York')
plt.ylabel('Average Number of Availability')
plt.xlabel('City')

plt.tight_layout()
plt.show()
```
